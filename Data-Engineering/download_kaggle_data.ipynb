# Download HarborMind Dataset from Kaggle

Because the fully engineered Parquet files are too large for GitHub (~1.3GB total), they are hosted on Kaggle. 
This notebook provides a simple way to download the data directly into your local workspace.

**Dataset Link:** [LA/LB Maritime Trajectory & AIS Dataset (2023-2025)](https://www.kaggle.com/datasets/4ff147dd68480c9e19df9e2f4d6eda1c3e41477b0cf56af0ac28f58b6eec2795)

### Prerequisites
You need a Kaggle API key (`kaggle.json`). 
1. Go to your Kaggle Account Settings.
2. Click **Create New Token** to download `kaggle.json`.


## 1. Download the Final ML-Ready Dataset


```python
import os
import kaggle

# Define the download directory
DOWNLOAD_DIR = "./data"
os.makedirs(DOWNLOAD_DIR, exist_ok=True)

# Dataset slug
DATASET_SLUG = "nguyenductuandat/lalb-ais-port-congestion-62023-122025"

print("Downloading ais_2023_2025_clean.parquet...")
kaggle.api.dataset_download_file(
    dataset=DATASET_SLUG,
    file_name="ais_2023_2025_clean.parquet",
    path=DOWNLOAD_DIR,
    force=True,
    quiet=False
)

# Kaggle downloads files as .zip if they are large, so we need to unzip it
import zipfile
zip_path = os.path.join(DOWNLOAD_DIR, "ais_2023_2025_clean.parquet.zip")
if os.path.exists(zip_path):
    print("Extracting zip file...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(DOWNLOAD_DIR)
    os.remove(zip_path) # Clean up zip file

```

## 2. (Optional) Download Individual Half-Year Batches


```python
# List of available batch files
batches = [
    "ais_2023H2.parquet",
    "ais_2024H1.parquet",
    "ais_2024H2.parquet",
    "ais_2025H1.parquet",
    "ais_2025H2.parquet"
]

print("Downloading half-year batches...")
for batch in batches:
    print(f"Fetching {batch}...")
    kaggle.api.dataset_download_file(
        dataset=DATASET_SLUG,
        file_name=batch,
        path=DOWNLOAD_DIR,
        force=True,
        quiet=True # Set to False to see progress bar
    )
    
    # Handle zip extraction if necessary
    zip_path = os.path.join(DOWNLOAD_DIR, f"{batch}.zip")
    if os.path.exists(zip_path):
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(DOWNLOAD_DIR)
        os.remove(zip_path)

```
